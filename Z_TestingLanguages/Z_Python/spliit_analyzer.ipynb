```python
import json

# Accessing top-level elements
trip_id = data['id']
trip_name = data['name']
currency = data['currency']

print(f"Trip ID: {trip_id}")
print(f"Trip Name: {trip_name}")
print(f"Currency: {currency}")


# Accessing expenses
expenses = data['expenses']
for expense in expenses:
    expense_date = expense['expenseDate']
    title = expense['title']
    category_grouping = expense['category']['grouping']
    category_name = expense['category']['name']
    amount = expense['amount']
    paid_by_id = expense['paidById']
    paid_for = expense['paidFor']  # This is a list of dictionaries
    is_reimbursement = expense['isReimbursement']
    split_mode = expense['splitMode']

    print(f"\nExpense Date: {expense_date}")
    print(f"Title: {title}")
    print(f"Category: {category_grouping} - {category_name}")
    print(f"Amount: {amount}")
    print(f"Paid by ID: {paid_by_id}")
    print("Paid for:")
    for person in paid_for:
        participant_id = person['participantId']
        shares = person['shares']
        print(f"  Participant ID: {participant_id}, Shares: {shares}")
    print(f"Is Reimbursement: {is_reimbursement}")
    print(f"Split Mode: {split_mode}")

# Accessing participants
participants = data['participants']
for participant in participants:
    participant_id = participant['id']
    participant_name = participant['name']
    print(f"\nParticipant ID: {participant_id}")
    print(f"Participant Name: {participant_name}")


# Example: Calculating total expenses
total_expenses = sum(expense['amount'] for expense in expenses)
print(f"\nTotal Expenses: {total_expenses}")

#Example: finding expenses in a certain category
hotel_expenses = [expense for expense in expenses if expense['category']['name'] == 'Hotel']

print("\nHotel Expenses")
for expense in hotel_expenses:
    print(expense['title'], expense['amount'])
```

```python
import json
import pandas as pd
from collections import defaultdict

with open('Spliit Export - 2025-02-01.json', 'r') as f:
    data = json.load(f)

expenses_by_category = defaultdict(int)

for expense in data['expenses']:
    category_name = expense['category']['name']
    amount = expense['amount']
    expenses_by_category[category_name] += amount

# Convert to pandas DataFrame
df = pd.DataFrame(list(expenses_by_category.items()), columns=['Category', 'Total Expenses'])

print(df)


#If you want to group by ID and category, you can do this:

expenses_by_id_and_category = defaultdict(lambda: defaultdict(int))

for expense in data['expenses']:
    paid_by_id = expense['paidById']
    category_name = expense['category']['name']
    amount = expense['amount']

    expenses_by_id_and_category[paid_by_id][category_name] += amount

# Convert to a list of dictionaries for DataFrame creation
data_for_df = []
for paid_by_id, categories in expenses_by_id_and_category.items():
    for category_name, total_amount in categories.items():
        data_for_df.append({'Paid By': paid_by_id, 'Category': category_name, 'Total Expenses': total_amount})

df_id_cat = pd.DataFrame(data_for_df)

print("\nDataFrame Grouped by ID and Category:")
print(df_id_cat)


#If you want to keep all the expense details in the dataframe:

expenses_list = []
for expense in data['expenses']:
    expenses_list.append({
        'Expense Date': expense['expenseDate'],
        'Title': expense['title'],
        'Category': expense['category']['name'],
        'Grouping': expense['category']['grouping'],
        'Amount': expense['amount'],
        'Paid By': expense['paidById'],
        # ... other fields as needed
    })

df_all_expenses = pd.DataFrame(expenses_list)
print("\nDataFrame with all expenses:")
print(df_all_expenses)
```

```python
import json
import pandas as pd

with open('Spliit Export - 2025-02-01.json', 'r') as f:
    data = json.load(f)

expenses_list = []
for expense in data['expenses']:
    expenses_list.append({
        'Paid By': expense['paidById'],
        'Amount': expense['amount']
    })

df = pd.DataFrame(expenses_list)

# Group by 'Paid By' and sum the 'Amount'
total_paid_by_each = df.groupby('Paid By')['Amount'].sum().reset_index()

print(total_paid_by_each)


#If you want to merge it with the participants names:

participants_dict = {p['id']: p['name'] for p in data['participants']}

total_paid_by_each['Payer Name'] = total_paid_by_each['Paid By'].map(participants_dict)

print("\nWith Payer Names:")
print(total_paid_by_each)
```

```python
import json
import pandas as pd

with open('Spliit Export - 2025-02-01.json', 'r') as f:
    data = json.load(f)

expenses_list = []
for expense in data['expenses']:
    for paid_for in expense['paidFor']:
        expenses_list.append({
            'Paid By': expense['paidById'],
            'Paid For': paid_for['participantId'],  # Added Paid For
            'Amount': expense['amount'] * (paid_for['shares'] / 100) if paid_for['shares'] > 0 else 0 # Amount divided by shares, handle 0 shares
        })

df = pd.DataFrame(expenses_list)


total_paid_by_each = df.groupby('Paid By')['Amount'].sum().reset_index()

participants_dict = {p['id']: p['name'] for p in data['participants']}
total_paid_by_each['Payer Name'] = total_paid_by_each['Paid By'].map(participants_dict)


print(total_paid_by_each)

#Total amount spent by each participant (what they paid for):
total_spent_by_each = df.groupby('Paid For')['Amount'].sum().reset_index()
total_spent_by_each['Participant Name'] = total_spent_by_each['Paid For'].map(participants_dict)

print("\nTotal spent by each participant:")
print(total_spent_by_each)
```
