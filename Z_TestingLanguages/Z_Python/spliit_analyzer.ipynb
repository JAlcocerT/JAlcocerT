```python
import json

# Function to read the JSON file
def read_json(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
    return data

# Function to calculate the payment share for each participant
def calculate_shares(expenses):
    # Create a dictionary to store total paid by each user
    total_paid_by_user = {}
    total_shares = {}

    for expense in expenses:
        # Calculate the total shares for the expense
        total_shares_expense = sum(participant['shares'] for participant in expense['paidFor'])

        # Calculate the amount each user should pay (total amount / total shares)
        amount_per_share = expense['amount'] / total_shares_expense

        # Track total paid for each user
        for participant in expense['paidFor']:
            participant_id = participant['participantId']
            shares = participant['shares']

            # Track total amount the user should pay
            if participant_id not in total_shares:
                total_shares[participant_id] = 0
            total_shares[participant_id] += shares * amount_per_share

            # Track how much the user paid for this expense
            if participant_id not in total_paid_by_user:
                total_paid_by_user[participant_id] = 0
            if expense['paidById'] == participant_id:
                total_paid_by_user[participant_id] += expense['amount']

    return total_paid_by_user, total_shares

# Function to calculate reimbursement or amount due
def calculate_reimbursement(total_paid_by_user, total_shares):
    reimbursement = {}
    for user_id in total_paid_by_user:
        paid = total_paid_by_user[user_id]
        owed = total_shares[user_id]
        if paid < owed:
            reimbursement[user_id] = owed - paid  # User owes this amount
        elif paid > owed:
            reimbursement[user_id] = -(paid - owed)  # User is owed this amount
    return reimbursement

# Main logic to read and process data
def main(filename):
    # Read data from the JSON file
    data = read_json(filename)

    # Calculate total paid and total owed for each user
    total_paid_by_user, total_shares = calculate_shares(data['expenses'])

    # Calculate the reimbursement amounts
    reimbursement = calculate_reimbursement(total_paid_by_user, total_shares)

    # Print the results
    print("Total Paid by User:")
    for user_id, paid in total_paid_by_user.items():
        print(f"{user_id}: {paid} {data['currency']}")

    print("\nTotal Owed by User:")
    for user_id, owed in total_shares.items():
        print(f"{user_id}: {owed} {data['currency']}")

    print("\nReimbursement (positive means user needs to pay, negative means user is owed money):")
    for user_id, amount in reimbursement.items():
        if amount != 0:
            print(f"{user_id}: {amount} {data['currency']}")

# Run the script
if __name__ == '__main__':
    filename = 'Spliit Export - 2025-02-01.json'  # Replace with your actual file path
    main(filename)
```

```python
import json
import pandas as pd
from collections import defaultdict

# Function to read the JSON file
def read_json(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
    return data

# Function to calculate and accumulate the total amount paid by each participant across all expenses per category
def calculate_participant_payments(expenses):
    # Dictionary to store payments by participant and category
    participant_totals = defaultdict(lambda: defaultdict(float))

    # Loop through each expense in the list
    for expense in expenses:
        category_name = expense['category']['name']  # Extract the category name
        first_participant = expense['paidFor'][0]  # Assuming first participant has the shares
        first_participant_share = first_participant['shares']
        total_amount = expense['amount'] / first_participant_share
        
        total_shares = sum(participant['shares'] for participant in expense['paidFor'])

        # Loop through each participant to calculate their share of the payment for this expense
        for participant in expense['paidFor']:
            participant_id = participant['participantId']
            shares = participant['shares']
            percentage_share = (shares / total_shares)  # Calculate the percentage of the total share
            participant_paid = total_amount * percentage_share

            # Add the amount paid by this participant to their running total, per category
            participant_totals[participant_id][category_name] += participant_paid

            # Print details of the expense and payment breakdown
            print(f"Expense Title: {expense['title']} for {total_amount}, Category: {category_name}, "
                  f"Participant ID: {participant_id}, Shares: {shares}, "
                  f"Percentage Share: {percentage_share:.2f}, Paid: {participant_paid:.2f}")
            
    return participant_totals

# Function to display the total amount paid by each participant across all categories
def display_total_paid(participant_totals):
    print("\nTotal Paid by Each Participant Across All Categories:")
    
    for participant_id, categories in participant_totals.items():
        total_paid = sum(categories.values())  # Sum the payments across all categories
        print(f"Participant ID: {participant_id}, Total Paid: {total_paid:.2f}")
        for category, paid in categories.items():
            print(f"  Category: {category}, Paid: {paid:.2f}")

# Function to convert the participant totals to a DataFrame
def convert_to_dataframe(participant_totals):
    # Convert the dictionary to a pandas DataFrame
    data = []
    for participant_id, categories in participant_totals.items():
        row = {'Participant ID': participant_id}
        row.update(categories)  # Add category-specific totals as columns
        data.append(row)

    df = pd.DataFrame(data)
    
    # Optionally, you can format all numeric columns to two decimal places
    df = df.round(2)
    
    return df

# Main logic to read and process data
def main(filename):
    # Read data from the JSON file
    data = read_json(filename)

    # Calculate the total paid by each participant across all categories
    participant_totals = calculate_participant_payments(data['expenses'])

    # Convert the data into a DataFrame for better viewing
    df_participant_totals = convert_to_dataframe(participant_totals)

    # Write the DataFrame to a CSV file
    df_participant_totals.to_csv('participant_totals_by_category.csv', index=False)

    # Display the total paid by each participant, including breakdown by category
    display_total_paid(participant_totals)

# Run the script
if __name__ == '__main__':
    filename = 'Spliit Export - 2025-02-01.json'  # Replace with your actual file path

    # Run the main logic
    main(filename)
```
